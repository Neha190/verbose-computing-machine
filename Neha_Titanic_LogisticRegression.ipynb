# DATA WRANGLING

```python
titanic.isnull()
```

```python
titanic.isnull().sum()
```

```python
titanic.drop('Cabin', axis= 1, inplace=True)
```

```python
titanic.head(15)
```

```python
sn.boxplot(x="Pclass",y="Age", data=titanic)
```

```python
sn.boxplot(x="Pclass",y="Fare", data=titanic)
```

```python
titanic.dropna(inplace=True)
```

```python
titanic.isnull().sum()
```

```python
sex=pd.get_dummies(titanic['Sex'],drop_first=True)
sex.head(5)
```

```python
embark=pd.get_dummies(titanic['Embarked'],drop_first=True)
embark.head(5)
```

```python
clas=pd.get_dummies(titanic['Pclass'],drop_first=True)
clas.head(5)
```

```python
titanic=pd.concat([titanic,sex,embark,clas],axis=1)
titanic.head(5)
```

```python
titanic.drop(['PassengerId','Pclass','Name','Sex','Ticket'],axis=1,inplace=True)
```

```python
titanic.head(5)
```

```python
titanic.drop(["Embarked"],axis=1,inplace=True)
titanic.head(5)
```

# TRAIN & TEST DATA

```python
X=titanic.drop("Survived",axis=1)
Y=titanic["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,Y_train)
```

```python
pred=logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

# ACCURACY CHECK

```python
accuracy_score(Y_test,pred)*100
```
